```python
import re
def clean_names(name):
    if isinstance(name,str):
        result = re.findall(r'[a-zA-z0-9 ]',name)
        res =''.join(result)
        return res
```

```python
zomato_df['name']=zomato_df['name'].apply(clean_names)
```

```python
import numpy as np
```

```python
def clean_rate(rate):
    if rate == "NEW" or rate == "-":
        return np.nan
    else:
        cleaned_rate =float(str(rate).split("/")[0])
        return cleaned_rate
zomato_df['rate']=zomato_df['rate'].apply(clean_rate)
```

```python
zomato_df['rate'].fillna(zomato_df['rate'].mean(),inplace=True)
```

```python
# find out top 10 resturants on the basis of voting.
zomato_df.groupby('name')[['votes']].mean().sort_values(by="votes",ascending=False).head(10)
```

```python
# find out top 10 resturants on the basis of rating.

zomato_df.groupby('name')[['rate']].mean().sort_values(by="rate",ascending=False).head(10)
```

```python
# Top 10 costliest Resturants in Banglore

top10_costliest_rest = zomato_df.groupby('name')[['approx_cost(for two people)']].mean().sort_values(by="approx_cost(for two people)",ascending=False).head(10)
top10_costliest_rest
```

```python
# Top 10 cheapest Resturants in Banglore

top10_cheapest_rest = zomato_df.groupby('name')[['approx_cost(for two people)']].mean().sort_values(by="approx_cost(for two people)",ascending=True).head(10)
top10_cheapest_rest
```

```python
# Resturant Outlets

outlets = zomato_df.groupby('name')[['rate']].count().sort_values('rate',ascending=False).reset_index().rename(columns={'rate':'outletcount'})
outlets
```

```python
# Location with Highest number of resturants

location_with_highest_rest = zomato_df.groupby('location')[['name']].count().sort_values('name',ascending=False).reset_index()
location_with_highest_rest
```

```python
# Top categories of Resturants in Banglore

top_categories_rest = zomato_df.groupby('rest_type')[['name']].count().sort_values('name',ascending=False).reset_index()
top_categories_rest
# OR

top_categories_rest1 = zomato_df.groupby('listed_in(type)')[['name']].count().sort_values('name',ascending=False).reset_index()
top_categories_rest1
```

```python
# Most Likes Cuisines

most_liked_cuisines = zomato_df.groupby('cuisines')[['votes']].count().sort_values('votes',ascending=False).head(10)
most_liked_cuisines
```

```python
# Number of Resturants respective to Cuisine Type

rest_cuisine_type = zomato_df.groupby('cuisines')[['name']].count().sort_values('name',ascending=False)
rest_cuisine_type
```

```python
# Most Expensive Cuisine with respective to Resturant Category

MostCostlyCuisineCategory = zomato_df.groupby(['cuisines','listed_in(type)'])[['approx_cost(for two people)']].mean().sort_values('approx_cost(for two people)',ascending=False)
MostCostlyCuisineCategory
```

```python
zomato_df['dish_liked']
```

```python
def splitData(data):
    return [x.strip() for x in str(data).split(',')]

zomato_df['dish_liked'].apply(splitData).explode('dish_liked').value_counts()[:10]
```

```python
# Top 10 dine out resturants

dine_out_rest = zomato_df[zomato_df['listed_in(type)'] == 'Dine-out'] \
    .groupby('name')[['rate']] \
    .mean() \
    .sort_values('rate', ascending=False)
dine_out_rest
```

```python
# Top 10 resturants for Desserts

dessert_rest = zomato_df[zomato_df['listed_in(type)'] == 'Desserts'].groupby('name')[['rate']].count().sort_values('rate', ascending=False).head(10)
dessert_rest
```

```python
# Dish Liked with respect to Lacation

DishLiked = zomato_df.groupby(['location','dish_liked'])[['name']].count().reset_index().sort_values('name', ascending=False)
DishLiked
```
